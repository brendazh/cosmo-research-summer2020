```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cl lensed from explanatory ini

ell = np.loadtxt("explanatory01_cl_lensed.dat",usecols=0)
clTT = np.loadtxt("explanatory01_cl_lensed.dat",usecols=1)

plt.xscale('log')
plt.xlabel(r'$\ell$')
plt.ylabel("TT")
plt.plot(ell, clTT)
plt.show
```

```python
# matter power spectrum from best fit parameters ini

k = np.loadtxt("best_fit_paper_z0_pk.dat",usecols=0)
P = np.loadtxt("best_fit_paper_z0_pk.dat",usecols=1)

plt.xscale('log')
plt.yscale('log')
plt.xlabel("k")
plt.ylabel("P")
plt.plot(k, P)
plt.show
```
